### Text Processing Functions:

```python
def process_text(text):
  """
  This function performs basic text cleaning (lowercasing)
  and extracts named entities (if needed).
  """
  doc = nlp(text)
  entities = [(ent.text, ent.label_) for ent in doc.ents]  # Extract entities and labels
  return doc.text_.lower(), entities  # No text cleaning here (performed in vectorization)

def vectorize_with_cleaning(texts):
  """
  This function cleans text further (removing punctuation and stop words)
  and tokenizes it.
  """
  # Convert text to lowercase and remove punctuation using pandas.Series.str.replace
  clean_texts = pd.Series(texts).str.lower().str.replace(f'[^{string.printable}]', '', regex=True)
  # Remove stop words using pre-built nltk.corpus stopwords
  from nltk.corpus import stopwords
  stop = stopwords.words('english')
  clean_texts = clean_texts.apply(lambda text: ' '.join([word for word in text.split() if word not in stop]))
  # Tokenization (vectorized using np.frompyfunc)
  tokens = np.frompyfunc(lambda x: x.split(), 1, 1)(clean_texts.to_numpy())
  return tokens

print("Text Processing Functions Defined!")
```
